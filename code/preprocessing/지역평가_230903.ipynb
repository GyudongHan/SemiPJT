```python
df1 = pd.read_csv('data/영세자영업+경영활동+현황(자치구별)_20230825012806.csv')
df1
```

```python
df1 = df1.iloc[:, 0:5]
df1 = df1.drop(1).reset_index(drop=True)
df1
```

```python
df1.columns = df1.iloc[0]
df1 = df1[1:].reset_index(drop=True)
df1
```

```python
df2 = pd.read_csv('data/영세자영업+자치구별+폐업+점포수_20230825012824.csv')
df2.columns = df2.iloc[0]
df2 = df2.drop(0).reset_index(drop=True)
df2
```

```python
df = df1.merge(df2, on = "자치구별(1)", how='inner')
df.rename(columns={'전체' : '폐업전체수'}, inplace = True)
```

```python
df.iloc[: , 1:10] = df.iloc[:, 1:10].astype(float)
df
```

```python
df['폐업비율'] = df['폐업전체수'] / df['운영점포수 (개)']
```

```python
df
```

```python
df['폐업비율순위'] = df['폐업비율'].rank()
df
```

```python
df['평균영업기간순위'] = df['평균영업기간 (년)'].rank(ascending = False)
df
```

```python
df['면적당매출액순위'] = df['면적당매출액 (백만원/3.3㎡)'].rank(ascending = False)
df
```

```python
df.iloc[: , 1:20] = df.iloc[:, 1:20].astype(float)
df
df['종합점수'] = df['폐업비율순위']+df['평균영업기간순위']+df['면적당매출액순위']
df = df.sort_values(by = '종합점수', ascending = False)
```
