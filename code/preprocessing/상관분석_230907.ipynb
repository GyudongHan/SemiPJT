```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.font_manager as fm

#font 오류 수정
font_list = fm.findSystemFonts()
font_name = None
for font in font_list:
    if 'NanumGothic' in font:
        font_name = fm.FontProperties(fname=font).get_name()
plt.rc('font', family=font_name)

# data importing
df = pd.read_csv('data/result/20230907_파생변수추가.csv')

# 분석할 컬럼 지정 
new_df = df.iloc[:, 5:]

# 상관분석
correlation_matrix = new_df.corr()
plt.figure(figsize=(15, 10)) # 잘 안보일 시 크기 조정
# annot = True 시 박스내에 상관계수가 보임.
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5) 
plt.title('상관 관계 히트맵')
plt.show()
```

```python
# 분석할 컬럼 지정 
name = '골목상권' # 분석할 유형 지정
new_df = df[df['상권_구분_코드_명'] == name]
target_df = new_df.iloc[:, 5:]

# 상관분석
correlation_matrix = target_df.corr()
plt.figure(figsize=(15, 10)) # 잘 안보일 시 크기 조정
# annot = True 시 박스내에 상관계수가 보임.
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5) 
plt.title(f'상관 관계 히트맵 ({name})')
plt.show()
```
