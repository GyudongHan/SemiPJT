```python
import numpy as np
import pandas as pd

# 2022년 1분기의 날짜 범위 생성 (날짜지정)
start_date = "2022-10-01"
end_date = "2022-12-31"
date_range = pd.date_range(start=start_date, end=end_date)

# 각 날짜에 해당하는 요일을 계산하여 새로운 열로 추가
df = pd.DataFrame({'날짜': date_range})
df['요일'] = df['날짜'].dt.day_name()

# 각 요일별로 카운트
day_counts = df['요일'].value_counts()

# 결과 출력
print(day_counts)
```

```python
import pandas as pd

# 분기별 날짜 범위 계산 함수
def calculate_quarter_dates(year, quarter):
    if quarter == 1:
        start_date = pd.Timestamp(f"{year}-01-01")
        end_date = pd.Timestamp(f"{year}-03-31")
    elif quarter == 2:
        start_date = pd.Timestamp(f"{year}-04-01")
        end_date = pd.Timestamp(f"{year}-06-30")
    elif quarter == 3:
        start_date = pd.Timestamp(f"{year}-07-01")
        end_date = pd.Timestamp(f"{year}-09-30")
    elif quarter == 4:
        start_date = pd.Timestamp(f"{year}-10-01")
        end_date = pd.Timestamp(f"{year}-12-31")
    else:
        return None, None

    return start_date, end_date

# 각 분기별 날짜 수 계산
year = 2020 # 연도 입력
for quarter in range(1, 5):
    start_date, end_date = calculate_quarter_dates(year, quarter)
    days_in_quarter = (end_date - start_date).days + 1
    print(f"{year}년 {quarter}분기는 {days_in_quarter}일입니다.")
```

```python
df['시간대_00~06_매출_금액_평균'] = df['시간대_00~06_매출_금액'] / df['점포수'] / df['일 수']
df['시간대_06~11_매출_금액_평균'] = df['시간대_06~11_매출_금액'] / df['점포수'] / df['일 수']
df['시간대_11~14_매출_금액_평균'] = df['시간대_11~14_매출_금액'] / df['점포수'] / df['일 수']
df['시간대_14~17_매출_금액_평균'] = df['시간대_14~17_매출_금액'] / df['점포수'] / df['일 수']
df['시간대_17~21_매출_금액_평균'] = df['시간대_17~21_매출_금액'] / df['점포수'] / df['일 수']
df['시간대_21~24_매출_금액_평균'] = df['시간대_21~24_매출_금액'] / df['점포수'] / df['일 수']
df.head()
```

```python
new_df = df.drop(df.columns[3:9], axis =1)
new_df = new_df.drop('일 수', axis =1)
new_df
new_df.to_csv('data/2020_2022_매출전처리최종.csv', encoding= 'utf-8 sig', index=False)
```
