```python
columns_to_sum = [
    '아파트_단지_수',
    '아파트_가격_1_억_미만_세대_수',
    '아파트_가격_1_억_세대_수',
    '아파트_가격_2_억_세대_수',
    '아파트_가격_3_억_세대_수',
    '아파트_가격_4_억_세대_수',
    '아파트_가격_5_억_세대_수',
    '아파트_가격_6_억_이상_세대_수'
]

for column in columns_to_sum:
    merged_df[column] = merged_df[column + '_x'] + merged_df[column + '_y']
    merged_df[column].fillna(merged_df[column + '_x'], inplace=True)
    merged_df.drop([column + '_x', column + '_y'], axis=1, inplace=True)
```
